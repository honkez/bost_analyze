```python
import matplotlib.pyplot as plt
import pandas as pd
import numpy as np
%matplotlib inline
```

```python
x = [0,1,2,3,4]
y2=[5,6,7,8,9]
y= [0,2,4,6,8]
#plt.plot(x,y,label='Urk',color='green',linestyle='-.',linewidth=4)
plt.figure(figsize=(8,5))
#Coolt sätt att ändra markörer till lines, finns dokumentation för detta
plt.plot(x,y,'b^-')
plt.plot(x,y2,label='Burk',marker='.')
plt.title('Skit diagram title', fontdict={'fontname':'Comic Sans MS','fontsize':20})
plt.legend()
plt.ylabel('Y axis')
plt.xlabel('X axis')
plt.xticks([0,1,2,3,4,5,6])
plt.yticks([0,2,4,6,8,10,12])
plt.show();
```
